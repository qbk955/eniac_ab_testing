#### In this case study, the objective is to conduct an A/B test on the "SHOP NOW" button on the main site of ENIAC Store to determine the most effective version. Following the test, a comprehensive statistical analysis will be performed to evaluate the significance of the results.



## Main questions i asked myself to prepare for the project

##### Inclusion of Variants (UX team prepared 4 versions)
###### We'll performe A/B on all of them because they are diffrent and it will increase chances to find best-performing one

##### Business Value
###### The most important thing is to improve CTR of the SHOP NOW button that can lead to higher user engagement and potentially higher sales

##### Main Metric to Measure Success
###### The main metric will be CTR 

##### Additional Metrics to Track
###### According to our A/B Test the best metrics would be "Drop-off rate for the linked page", "Homepage-return rate for the category pages" but worth tracking would also be "Time on Site" and "User Navigation Path"

##### Null and Alternative Hypotheses
###### Null Hypothesis (H0): There is no difference in the click-through rate (CTR) of the "SHOP NOW" button between the current version and any of the new variants (A, B, C, and D)
###### Alternative Hypothesis (H1): There is a difference in the click-through rate (CTR) of the "SHOP NOW" button between the current version and at least one of the new variants (A, B, C, and D).
###### Chi-square test need to be performed

#####  Threshold for Statistical Significance and Minimum Detectable Effect
###### I will set p-value of 0.05 - It's not high risk research for stakeholders and its standard value in this industry.
###### Minimum detectable effect 20% - which means an increase from a CTR of 0.02 (2%) to 0.024 (2.4%)

##### Development of the Experiment
###### This experiment could be developed using external tools such as A/B Tasty, Optimizely, or Google Optimize, which are specifically designed for running A/B tests and multivariate tests

#### Summary
##### To improve the "SHOP NOW" button, I would run an A/B test including all suggested variants (A, B, C, and D). The main metric to measure success would be the CTR, with additional metrics such as conversion rate, bounce rate, time on site, and navigation path also tracked. The null hypothesis would state no difference in CTR between variants, with a 0.05 significance level set for statistical testing. The minimum detectable effect would be a 0.4 percentage point increase in CTR. The experiment could be implemented using external A/B testing tools. This approach aims to identify the most effective design for the "SHOP NOW" button, enhancing user engagement and driving business value.
